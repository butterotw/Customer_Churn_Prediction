# Streamlit (Page 1 + Page 2) — Data Overview + Association Rule Mining

This notebook generates a **working Streamlit app (`app.py`)** for:

- **Page 1: Data Overview / EDA**
  - Show dataframe preview + summary
  - Charts
  - Sidebar: exclude numerical / categorical columns (tutorial style)

- **Page 2: Association Rule Mining (ARM)**
  - Show frequent itemsets + association rules
  - Sidebar: thresholds (min_support, min_confidence, min_lift, top_n)
  - Includes **speed controls** to avoid FP-Growth explosion

✅ Suggested structure:
- `train_models.ipynb` → training + saving models (later)
- `features_eda_arm.ipynb` (this notebook) → EDA + ARM + generating Streamlit app
- `app.py` → Streamlit UI

---

## How to use
1. Run the cells to generate `app.py`.
2. Run in terminal:
   - `streamlit run app.py`

````
import os

CSV_PATH = "BankChurners_clean.csv"
# If your CSV is elsewhere, set absolute path:
# CSV_PATH = r"C:\Users\...\BankChurners_clean.csv"

assert os.path.exists(CSV_PATH), f"Cannot find {CSV_PATH}. Put it next to this notebook or set CSV_PATH to absolute path."
print("✅ Found CSV:", os.path.abspath(CSV_PATH))
````

````
# Write Streamlit app to app.py
import os

app_code = '\nimport streamlit as st\nimport pandas as pd\nimport numpy as np\nimport matplotlib.pyplot as plt\nfrom mlxtend.frequent_patterns import fpgrowth, association_rules\n\n# -----------------------------\n# Page config (MUST be first Streamlit command)\n# -----------------------------\nst.set_page_config(page_title="Bank Churn — EDA + ARM", layout="wide")\n\nDATA_PATH = "BankChurners_clean.csv"\n\n# -----------------------------\n# Data loading\n# -----------------------------\n@st.cache_data\ndef load_data(path: str) -> pd.DataFrame:\n    df = pd.read_csv(path)\n    df = df.drop(columns=["CLIENTNUM"], errors="ignore")\n    df["target"] = df["Attrition_Flag"].map({"Existing Customer": 0, "Attrited Customer": 1})\n    return df\n\ndf = load_data(DATA_PATH)\n\n# Identify columns\nall_num = df.select_dtypes(include=["int64", "float64", "int32", "float32"]).columns.tolist()\nall_cat = df.select_dtypes(exclude=["int64", "float64", "int32", "float32"]).columns.tolist()\nfor col in ["target"]:\n    if col in all_num: all_num.remove(col)\n    if col in all_cat: all_cat.remove(col)\nif "Attrition_Flag" in all_cat:\n    all_cat.remove("Attrition_Flag")\n\n# -----------------------------\n# Sidebar: page selection\n# -----------------------------\nst.sidebar.subheader("Select Page")\npage = st.sidebar.selectbox(\n    "Options",\n    ["1. Data Overview (EDA)", "2. Association Rule Mining (ARM)"]\n)\n\n# -----------------------------\n# Page 1 sidebar: exclusions\n# -----------------------------\nst.sidebar.subheader("EDA: Attribute Selection")\nexclude_num = st.sidebar.multiselect("Exclude numerical columns", all_num, default=[])\nexclude_cat = st.sidebar.multiselect("Exclude categorical columns", all_cat, default=[])\n\n# -----------------------------\n# Page 2 sidebar: ARM thresholds + speed controls\n# -----------------------------\nst.sidebar.subheader("ARM Thresholds")\nmin_support = st.sidebar.slider("Minimum Support", 0.01, 0.50, 0.02, 0.01)\nmin_confidence = st.sidebar.slider("Minimum Confidence", 0.10, 1.0, 0.60, 0.05)\nmin_lift = st.sidebar.slider("Minimum Lift", 0.50, 10.0, 1.00, 0.10)\ntop_n = st.sidebar.slider("Top N Rules", 10, 200, 50, 10)\n\nst.sidebar.subheader("ARM Speed Control (Recommended)")\nmax_items = st.sidebar.slider("Keep Top-N frequent items", 20, 200, 80, 10)\nmax_len = st.sidebar.slider("Max itemset length", 2, 5, 3, 1)\n\n# -----------------------------\n# Helpers\n# -----------------------------\ndef apply_exclusions(df_in: pd.DataFrame, exc_num: list, exc_cat: list) -> pd.DataFrame:\n    df2 = df_in.copy()\n    drop_cols = [c for c in (exc_num + exc_cat) if c in df2.columns]\n    return df2.drop(columns=drop_cols, errors="ignore")\n\ndef plot_target_distribution(df_in: pd.DataFrame):\n    counts = df_in["target"].value_counts().sort_index()\n    fig, ax = plt.subplots()\n    ax.bar(["0 (Existing)", "1 (Attrited)"], counts.values)\n    ax.set_title("Target Distribution")\n    ax.set_ylabel("Count")\n    st.pyplot(fig)\n\ndef plot_numeric_hist(df_in: pd.DataFrame, col: str):\n    fig, ax = plt.subplots()\n    ax.hist(df_in[col].dropna(), bins=30)\n    ax.set_title(f"Histogram — {col}")\n    ax.set_xlabel(col)\n    ax.set_ylabel("Frequency")\n    st.pyplot(fig)\n\ndef plot_cat_counts(df_in: pd.DataFrame, col: str):\n    vc = df_in[col].astype(str).value_counts().head(20)\n    fig, ax = plt.subplots()\n    ax.bar(vc.index, vc.values)\n    ax.set_title(f"Top categories — {col}")\n    ax.set_xlabel(col)\n    ax.set_ylabel("Count")\n    plt.xticks(rotation=45, ha="right")\n    st.pyplot(fig)\n\ndef build_arm_dataset(df_in: pd.DataFrame, cols_for_arm: list) -> pd.DataFrame:\n    """\n    Tutorial-style: use selected prepared categorical/binned features.\n    One-hot encode and mine rules.\n    """\n    work = df_in[cols_for_arm].dropna().copy()\n    work = pd.get_dummies(work)\n    work = work.astype(bool)\n    return work\n\ndef run_fpgrowth_arm(tx: pd.DataFrame, min_support: float, min_confidence: float, min_lift: float,\n                    top_n: int, max_items: int, max_len: int):\n    # Keep only top-N frequent items to avoid explosion\n    item_freq = tx.mean(axis=0).sort_values(ascending=False)\n    keep_cols = item_freq.head(max_items).index\n    tx2 = tx[keep_cols].copy()\n\n    itemsets = fpgrowth(tx2, min_support=min_support, use_colnames=True, max_len=max_len)\n    if itemsets.empty:\n        return itemsets, pd.DataFrame()\n\n    rules = association_rules(itemsets, metric="confidence", min_threshold=min_confidence)\n    if rules.empty:\n        return itemsets, rules\n\n    rules = rules[rules["lift"] >= min_lift].copy()\n    rules.sort_values(["lift", "confidence", "support"], ascending=False, inplace=True)\n    rules = rules.head(top_n).reset_index(drop=True)\n\n    rules["antecedents"] = rules["antecedents"].apply(lambda s: ", ".join(list(s)))\n    rules["consequents"] = rules["consequents"].apply(lambda s: ", ".join(list(s)))\n\n    return itemsets, rules[["antecedents","consequents","support","confidence","lift"]]\n\n\n# -----------------------------\n# Page 1: EDA\n# -----------------------------\nif page == "1. Data Overview (EDA)":\n    st.title("Page 1 — Data Overview & EDA")\n\n    df_eda = apply_exclusions(df, exclude_num, exclude_cat)\n\n    c1, c2, c3 = st.columns(3)\n    c1.metric("Rows", df_eda.shape[0])\n    c2.metric("Columns", df_eda.shape[1])\n    c3.metric("Attrition Rate (%)", f"{df[\'target\'].mean()*100:.2f}")\n\n    st.subheader("DataFrame Preview")\n    st.dataframe(df_eda.head(50), use_container_width=True)\n\n    st.subheader("Target Distribution")\n    plot_target_distribution(df)\n\n    st.subheader("Summary Statistics (Numerical)")\n    num_cols = df_eda.select_dtypes(include=["int64","float64","int32","float32"]).columns.tolist()\n    if "target" in num_cols:\n        num_cols.remove("target")\n    if len(num_cols) == 0:\n        st.warning("No numerical columns after exclusions.")\n    else:\n        st.dataframe(df_eda[num_cols].describe().T, use_container_width=True)\n\n        st.subheader("Numeric Histograms")\n        pick_num = st.multiselect("Select numeric columns", num_cols, default=num_cols[:4])\n        for col in pick_num:\n            plot_numeric_hist(df_eda, col)\n\n    st.subheader("Categorical Distributions")\n    cat_cols = df_eda.select_dtypes(exclude=["int64","float64","int32","float32"]).columns.tolist()\n    cat_cols = [c for c in cat_cols if c != "Attrition_Flag"]\n    if len(cat_cols) == 0:\n        st.warning("No categorical columns after exclusions.")\n    else:\n        pick_cat = st.multiselect("Select categorical columns", cat_cols, default=cat_cols[:3])\n        for col in pick_cat:\n            plot_cat_counts(df_eda, col)\n\n\n# -----------------------------\n# Page 2: ARM\n# -----------------------------\nelif page == "2. Association Rule Mining (ARM)":\n    st.title("Page 2 — Association Rule Mining (ARM)")\n    st.write("Runs **FP-Growth** + association rules. Use sidebar thresholds like the tutorial.")\n\n    # Default ARM columns based on your DataMiningProject.ipynb\n    default_arm_cols = [\n        "Age_Group","Education_Level","Marital_Status","Income_Category",\n        "Utilization_Level","Transaction_Amount_Level","Transaction_Count_Level",\n        "Inactivity_Level","Attrition_Flag"\n    ]\n    available_default = [c for c in default_arm_cols if c in df.columns]\n\n    cols_for_arm = st.multiselect(\n        "Select columns for ARM (tutorial-style features)",\n        options=[c for c in df.columns if c != "target"],\n        default=available_default\n    )\n\n    if len(cols_for_arm) < 2:\n        st.warning("Select at least 2 columns for ARM.")\n        st.stop()\n\n    if st.button("Run ARM"):\n        with st.spinner("Encoding transactions..."):\n            tx = build_arm_dataset(df, cols_for_arm)\n\n        st.caption(f"Transactions: {tx.shape[0]} rows × {tx.shape[1]} items (before top-N filter)")\n\n        with st.spinner("Mining frequent itemsets + rules..."):\n            itemsets, rules = run_fpgrowth_arm(\n                tx,\n                min_support=min_support,\n                min_confidence=min_confidence,\n                min_lift=min_lift,\n                top_n=top_n,\n                max_items=max_items,\n                max_len=max_len\n            )\n\n        st.subheader("Frequent Itemsets")\n        if itemsets.empty:\n            st.warning("No frequent itemsets found. Try lowering min_support OR increasing max_items.")\n        else:\n            st.dataframe(itemsets.sort_values("support", ascending=False).head(50), use_container_width=True)\n\n        st.subheader("Association Rules")\n        if rules.empty:\n            st.warning("No rules found. Try lowering min_confidence/min_lift or increasing max_items.")\n        else:\n            st.dataframe(rules, use_container_width=True)\n\n            st.subheader("Support vs Confidence")\n            fig, ax = plt.subplots()\n            ax.scatter(rules["support"], rules["confidence"])\n            ax.set_xlabel("Support")\n            ax.set_ylabel("Confidence")\n            ax.set_title("Top rules (Support vs Confidence)")\n            st.pyplot(fig)\n\n    else:\n        st.info("Set thresholds in the sidebar, then click **Run ARM**.")\n        st.stop()\n'

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("✅ Generated app.py at:", os.path.abspath("app.py"))
````

## Run Streamlit (Windows)

Run in PowerShell / CMD:

```powershell
streamlit run app.py
```
