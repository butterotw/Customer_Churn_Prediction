```python
df['Utilization_Level'] = pd.qcut(
    df['Avg_Utilization_Ratio'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Transaction_Amount_Level'] = pd.qcut(
    df['Total_Trans_Amt'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Transaction_Count_Level'] = pd.qcut(
    df['Total_Trans_Ct'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Inactivity_Level'] = pd.cut(
    df['Months_Inactive_12_mon'],
    bins=[-1, 1, 3, df['Months_Inactive_12_mon'].max()],
    labels=['Few', 'Moderate', 'Many']
)
```

```python
df['Age_Group'] = pd.cut(
    df['Customer_Age'],
    bins=[0, 40, 55, df['Customer_Age'].max()],
    labels=['Young', 'Middle', 'Old']
)
```

```python
df[['Utilization_Level',
    'Transaction_Amount_Level',
    'Transaction_Count_Level',
    'Inactivity_Level',
    'Age_Group']].value_counts()
```

```python
aprtiori_features = ['Age_Group','Education_Level','Marital_Status','Income_Category','Utilization_Level', 'Transaction_Amount_Level','Transaction_Count_Level','Inactivity_Level','Attrition_Flag']

aprtiori_features = df[aprtiori_features].dropna()
```

```python
df_encoded = pd.get_dummies(aprtiori_features)
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

frequent_itemsets = fpgrowth(
    df_encoded,
    min_support=0.02,
    use_colnames=True
)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric='confidence',
    min_threshold=0.6
)
```

```python
rules_attrition = rules[
    rules['consequents']
    .apply(lambda x: 'Attrition_Flag_Attrited Customer' in x)
]
```

```python
rules_attrition_sorted = rules_attrition.sort_values(
    by=['lift', 'confidence'],
    ascending=False
)

rules_attrition_sorted[['antecedents','consequents','support','confidence','lift']].head(15)
```

Random Forest


```python
# new df so previous doesnt get overwritten

random_forest_df = df[['Attrition_Flag','Avg_Utilization_Ratio','Total_Trans_Amt','Total_Trans_Ct','Months_Inactive_12_mon','Customer_Age','Gender','Education_Level','Income_Category','Marital_Status','Card_Category']].copy()
```

```python
#
random_forest_df['Attrition'] = (random_forest_df['Attrition_Flag'] == 'Attrited Customer').astype(int)
random_forest_df = random_forest_df.drop(columns=['Attrition_Flag'])

```

```python
y = random_forest_df['Attrition']
X = random_forest_df.drop(columns=['Attrition'])
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=32
)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

random_forest.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

random_forest_prob = random_forest.predict_proba(X_test)[:, 1]

roc_auc_rf = roc_auc_score(y_test, random_forest_prob)
print("Random Forest ROC-AUC:", roc_auc_rf)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, _ = roc_curve(y_test, random_forest_prob)
auc = roc_auc_score(y_test, random_forest_prob)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Random Forest')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

rf_preds = (random_forest_prob >= 0.5).astype(int) #0.5 is randomly chosen pick better val after writing

print("Confusion Matrix:")
print(confusion_matrix(y_test, rf_preds))

print("\nClassification Report:")
print(classification_report(y_test, rf_preds))
```

```python
confusion_mat = confusion_matrix(y_test, rf_preds)

plt.figure(figsize=(5, 4))
sns.heatmap(
    confusion_mat,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Not Attrite', 'Attrite'],
    yticklabels=['Not Attrite', 'Attrite']
)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion mat')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(
    random_forest_prob[y_test == 0],
    bins=30,
    color='blue',
    label='Not Attrite',
    stat='density',
    kde=True
)
sns.histplot(
    random_forest_prob[y_test == 1],
    bins=30,
    color='red',
    label='Attrite',
    stat='density',
    kde=True
)
plt.xlabel('Predicted Churn Probability')
plt.title('Probability Distribution')
plt.legend()
plt.show()
```

Naive Beyes


```python
naive_bayes_df = df[['Attrition_Flag','Avg_Utilization_Ratio','Total_Trans_Amt', 'Total_Trans_Ct','Months_Inactive_12_mon', 'Customer_Age', 'Gender', 'Education_Level','Income_Category','Marital_Status','Card_Category']].copy()
```

```python
naive_bayes_df['Attrition'] = (naive_bayes_df['Attrition_Flag'] == 'Attrited Customer').astype(int)

naive_bayes_df = naive_bayes_df.drop(columns=['Attrition_Flag'])
```

```python
naive_Y=naive_bayes_df['Attrition']
naive_X=naive_bayes_df.drop(columns=['Attrition'])
```

```python

naive_X = pd.get_dummies(naive_X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train_nb, X_test_nb, y_train_nb, y_test_nb = train_test_split(
    naive_X,
    naive_Y,
    test_size=0.2,
    stratify=naive_Y,
    random_state=42
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_nb = MinMaxScaler()

X_train_nb = scaler_nb.fit_transform(X_train_nb)
X_test_nb = scaler_nb.transform(X_test_nb)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train_nb, y_train_nb)
```

```python

nb_probs = nb.predict_proba(X_test_nb)[:, 1]
print("Naive Bayes ROC-AUC:", roc_auc_score(y_test_nb, nb_probs))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr_nb, tpr_nb, _ = roc_curve(y_test_nb, nb_probs)
auc_nb = roc_auc_score(y_test_nb, nb_probs)

plt.figure(figsize=(6, 5))
plt.plot(fpr_nb, tpr_nb, label=f'Naive Bayes (AUC = {auc_nb:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Naive Bayes')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python

nb_preds = (nb_probs >= 0.5).astype(int)
cm_nb = confusion_matrix(y_test_nb, nb_preds)

plt.figure(figsize=(5, 4))
sns.heatmap(
    cm_nb,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Not Attrite', 'Attrite'],
    yticklabels=['Not Attrite', 'Attrite']
)
plt.title('Confusion Matrix – Naive Bayes')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```
